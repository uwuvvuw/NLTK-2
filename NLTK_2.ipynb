Write a function that can count (1) the number of sentences in any paragraph of text given as a parameter and print (2) each sentence separately.

Copy the function into the submission sheet and screenshot your output

```python
# Write your code here, execute the function and leave the output before saving the ipynb file

# count the number of sentences in a paragraph of text
def count_sentences(paragraph):
  # a string containing a paragraph of text
  sentences = paragraph.split("\n")

  # the number of sentences in the paragraph
  return len(sentences)

def main():
  # get a paragraph of text from the user
  paragraph = "Stroke occurs when the blood supply to part of the brain is interrupted, depriving brain cells of oxygen and nutrients.\nThis can cause brain damage and, in some cases, death.\nStroke is the fifth leading cause of death in the United States and a leading cause of disability."

  # count the number of sentences in the paragraph
  number_of_sentences = count_sentences(paragraph)

  # print the number of sentences in the paragraph
  print("The paragraph contains {} sentences.\n".format(number_of_sentences))

  # print the output as a list of numbered sentences
  sentences = paragraph.split("\n")
  for i in range(1, number_of_sentences + 1):
    print(i, "-", sentences[i - 1])

if __name__ == "__main__":
  main()
```

Find out what each POS tag represents when the sentence below is analysed by the NLTK POS tagger.

***I have a dream that my four little children will one day live in a nation
where they will not be judged by the color of their skin but by the content of their character***

Edit/add to the function such that it prints out only words that are tagged as verbs, pronouns, and numbers SEPERATELY

```python
# Write your code here, execute the function and leave the output before saving the ipynb file
import nltk
nltk.download('averaged_perceptron_tagger')
text = nltk.word_tokenize("I have a dream that my four little children will one day live in a nation where they will not be judged by the color of their skin but by the content of their character")

# get the part-of-speech tags for the words in the text
tags = nltk.pos_tag(text)

# create a list of verbs, pronouns, and numbers
verbs = []
pronouns = []
numbers = []
for tag in tags:
  if tag[1] == "VB":
    verbs.append(tag[0])
  elif tag[1] == "PRP":
    pronouns.append(tag[0])
  elif tag[1] == "CD":
    numbers.append(tag[0])

# print the lists of verbs, pronouns, and numbers
print("Verbs:", verbs)
print("Pronouns:", pronouns)
print("Numbers:", numbers)
```

```python
# Write your code here, execute the function and leave the output before saving the ipynb file
name_text = open("Natasha.txt", "r")

# create an empty dictionary
d = dict()

# loop through each line of the file
for line in name_text:
    # remove the leading spaces and newline character
    line = line.strip()

    # convert the characters in line to
    # lowercase to avoid case mismatch
    line = line.lower()

    # split the line into words
    words = line.split(" ")


    # iterate over each word in line
    for word in words:
        # check if the word is already in dictionary
        if word in d:
            # increment count of word by 1
            d[word] = d[word] + 1
        else:
            # add the word to dictionary with count 1
            d[word] = 1

# print the contents of dictionary
for key in list(d.keys()):
    print(key, ":", d[key])
```
